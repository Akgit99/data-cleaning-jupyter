```python
# top 10 populous countries in 2023
top_5_populous = df_countries.sort_values(by='2023', ascending=False).head(10)
top_5_populous[['Country Name', '2023']]
```

```python
sns.boxplot(x='Country Name', y='2023', data=top_5_populous)
plt.title('Population Distribution in 2023 by Country')
plt.xticks(rotation=90)
```

```python
sns.pairplot(top_5_populous[['1960', '1970', '1980', '1990', '2000', '2010', '2023']])
```

```python
pivot_df = top_5_populous.pivot_table(index='Country Name', columns='Year', values='Population', aggfunc='sum')
pivot_df.head()
```
